```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
B_clim = np.load('outputs/Beijingclimate.npy')
B_loss = np.load('outputs/Beijingloss.npy')
B_pred = np.load('outputs/Beijingpreds.npy')
B_gt = np.load('outputs/Beijingtruth.npy')
K_clim = np.load('outputs/Knoxvilleclimate.npy')
K_loss = np.load('outputs/Knoxvilleloss.npy')
K_pred = np.load('outputs/Knoxvillepreds.npy')
K_gt = np.load('outputs/Knoxvilletruth.npy')
N_clim = np.load('outputs/Nairobiclimate.npy')
N_loss = np.load('outputs/Nairobiloss.npy')
N_pred = np.load('outputs/Nairobipreds.npy')
N_gt = np.load('outputs/Nairobitruth.npy')
P_clim = np.load('outputs/Parisclimate.npy')
P_loss = np.load('outputs/Parisloss.npy')
P_pred = np.load('outputs/Parispreds.npy')
P_gt = np.load('outputs/Paristruth.npy')
R_clim = np.load('outputs/Riyadhclimate.npy')
R_loss = np.load('outputs/Riyadhloss.npy')
R_pred = np.load('outputs/Riyadhpreds.npy')
R_gt = np.load('outputs/Riyadhtruth.npy')
S_clim = np.load('outputs/Sydneyclimate.npy')
S_loss = np.load('outputs/Sydneyloss.npy')
S_pred = np.load('outputs/Sydneypreds.npy')
S_gt = np.load('outputs/Sydneytruth.npy')
```

```python
plt.plot(B_pred[:,7], color = 'r', label = 'prediction')
plt.plot(B_clim[0,7,:], color = 'g', label = 'climate')
plt.plot(B_gt[:,7], color = 'k',label = 'truth')
plt.legend()
```

```python
fig, ax = plt.subplots(2,3, figsize = (16, 10))
ax[0][0].plot(B_loss)
ax[0][0].set_yscale('log')
ax[0][0].set_title('Beijing')
ax[0][0].set_xlabel('Epoch')
ax[0][0].set_ylabel('loss')
ax[0][1].plot(K_loss)
ax[0][1].set_yscale('log')
ax[0][1].set_title('Knoxville')
ax[0][1].set_xlabel('Epoch')
ax[0][1].set_ylabel('loss')
ax[0][2].plot(N_loss)
ax[0][2].set_yscale('log')
ax[0][2].set_title('Nairobi')
ax[0][2].set_xlabel('Epoch')
ax[0][2].set_ylabel('loss')
ax[1][0].plot(P_loss)
ax[1][0].set_yscale('log')
ax[1][0].set_title('Paris')
ax[1][0].set_xlabel('Epoch')
ax[1][0].set_ylabel('loss')
ax[1][1].plot(R_loss)
ax[1][1].set_yscale('log')
ax[1][1].set_title('Riyadh')
ax[1][1].set_xlabel('Epoch')
ax[1][1].set_ylabel('loss')
ax[1][2].plot(S_loss)
ax[1][2].set_yscale('log')
ax[1][2].set_title('Sydney')
ax[1][2].set_xlabel('Epoch')
ax[1][2].set_ylabel('loss')
fig.suptitle('Loss-Epoch charts for 6 RNNs', fontsize = 24)
```

```python
titles = ['Climate Water Deficit', 'Palmer Drought Severity Index', 'Precipitation', 'Soil Moisture', 
         'Snow Water Equivalent', 'Downward Shortwave Radiation', 'Vapor Pressure', 'Wind Speed']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
fig, ax = plt.subplots(4,2, figsize = (16, 24))
for i in range(8):
    col = i%2
    row = i//2
    if (row == 0) and (col == 0):
        ax[row][col].plot(B_pred[:,i], color = 'r', label = 'prediction')
        ax[row][col].plot(B_clim[0,i,:], color = 'g', label = 'climate')
        ax[row][col].plot(B_gt[:,i], color = 'k',label = 'truth')
    else:
        ax[row][col].plot(B_pred[:,i], color = 'r')
        ax[row][col].plot(B_clim[0,i,:], color = 'g')
        ax[row][col].plot(B_gt[:,i], color = 'k')
    ax[row][col].set_xticks(np.arange(12))
    ax[row][col].set_xticklabels(months)
    ax[row][col].set_title(titles[i])
fig.suptitle('Beijing RNN outputs (2017)', fontsize = 20, y = .92)
fig.legend(loc = (.92,.92))
plt.show()
```
